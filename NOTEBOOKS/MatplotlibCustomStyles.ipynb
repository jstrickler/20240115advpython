# Custom Matplotlib styles
You can put custom Matplotlib style settings in a file (extension `.mplstyle`) and load it with `mpl.style.use()` or `mpl.style.context()`. The name before the extension is the style name (e.g., for `datasci.mplstyle`, the style name is "datasci". 

A complete list of settings: https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.rcParams

<div class="alert alert-block alert-warning">
<b>WARNING:</b> Not all settings in that list can be used -- some are not applicable to plots and will be ignored. A list of those settings is at the end of this notebook.
</div>

## Where to put custom styles
The location for custom styles is the `stylelib` folder in the Matplotlib config folder. The `stylelib` folder is not created by default, so you may have to create it before you use your custom styles. You can put any number of custom style files in `stylelib`.

<div class="alert alert-block alert-info">
<b>TIP:</b>To find the config folder, call <code>mpl.get_configdir()</code>.  
</div>



```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import os
```

```python
config_dir = mpl.get_configdir()
config_dir
```

## Try it out
To make the following code work, copy `mycustom.mplstyle` from the top level of the student files to the `stylelib` folder, as described in the previous section.

`plt.style.use()` sets the style for all subsequent plots. To use a style for only one plot, use 

    with plt.style.context('mycustom'):
        # call plt.plot() here

```python
plt.style.use('mycustom')
plt.title("My Terrible Graph")
for i in range(1,5):
    data = np.random.random_sample((10,)) * 10
    plt.plot(data, label=f"Set {i}")  # add label for legend
    plt.xlabel("item")
    plt.ylabel("random data")
    plt.ylim(0, 15)  # make y axis longer so legend doesn't overlap plot
plt.legend()
```

## View the custom style settings
This code will show the contents of the custom styles file `mycustom.mplstyle`

```python
style_path = os.path.join(config_dir, 'stylelib', 'mycustom.mplstyle')
with open(style_path) as style_in:  # read and print contents of custom styles file
    print(style_in.read())
```

### Config parameters that are not related to style
The following parameters for Matplotlib are not related to styles, and will be ignored in a custom styles file. 


+ backend
+ backend_fallback
+ date.epoch
+ docstring.hardcopy
+ figure.max_open_warning
+ figure.raise_window
+ interactive
+ savefig.directory
+ timezone
+ tk.window_focus
+ toolbar
+ webagg.address
+ webagg.open_in_browser
+ webagg.port
+ webagg.port_retries
